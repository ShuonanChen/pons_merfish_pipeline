# how to test if each of them are in merfish and then save the results in pickle>

```python
import time
from pathlib import Path
from trimesh import load as load_trimesh
from multiprocessing import Pool, cpu_count
def process_chunk(chunk, mesh):
    return mesh.contains(points=chunk)

def check_chunk_in_mesh(mesh, coords):
    chunk_size = 1000
    n_points = coords.shape[0]
    chunks = [coords[i : i + chunk_size] for i in range(0, n_points, chunk_size)]
    with Pool(cpu_count()) as pool:
        results = pool.starmap(process_chunk, [(chunk, mesh) for chunk in chunks])
    inside = np.concatenate(results)
    return inside
```

```python
# around 2128 secs. (>30 min)
time1=time.time()
LC_region_cells = dict()
for name in slicename_all:
    withinmesh = []
    for mesh in allmeshes:
        withinmesh.append(check_chunk_in_mesh(mesh, reg_points_all[name].iloc[:,1:].values))
    LC_region_cells[name] = pd.DataFrame(np.array(withinmesh).T, columns = allmeshes_names)
    
print(f'totaltime needed: {time.time()-time1}')    
```
