```python
adata_mer.layers['counts'] = adata_mer.raw.X
scvi.model.SCVI.setup_anndata(adata_mer, batch_key='section_name', layer = 'counts')
model = scvi.model.SCVI(adata_mer)
model.train()
adata_mer.obsm['X_scvi'] = model.get_latent_representation()
```

```python
sc.pp.neighbors(adata_mer, use_rep='X_scvi')
sc.tl.leiden(adata_mer,resolution=.5)
sc.tl.umap(adata_mer)
```

```python
intersted_genes = ['Trhr','Pdyn','Tacr3','Shox2','Gpr101',]
fig, axes = plt.subplots(1,5,figsize=(35,5),)
for k in range(len(intersted_genes)):
    sc.pl.umap(adata_mer, ax = axes[k],color=intersted_genes[k], sort_order=True, vmax = 100 , show=False,
               size =np.clip(adata_mer.X[:,np.where(adata_mer.var_names==intersted_genes[k])[0][0]],.01,None),
               alpha =.9, cmap = 'Reds')
```
