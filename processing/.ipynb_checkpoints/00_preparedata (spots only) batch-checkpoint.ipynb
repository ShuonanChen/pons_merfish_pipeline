```python
import glob, sys, os
from pprint import pprint
import matplotlib.pyplot as plt
import numpy as np
import tqdm
import pandas as pd
import scanpy as sc
from func.prepare_img_utils import *
```

```python
globalscaling = 32
save_destination = '/home/shuonan.chen/scripts/Pons_MERFISH/analysis_pipeline/'  # should be cwd
main_dir = '/allen/aind/scratch/shuonan.chen/pons_images/'  # where raw images are 
all_tifs = (main_dir + '*/' + 'analyzed_data/*/region_*/images/*.tif')
name_of_slice_all = [[s for s in f.split('/') if "Mouse" in s][0] for f in glob.glob(all_tifs)]
```

```python
verbose=False
for name_of_slice in name_of_slice_all:
    thisdircheck = (main_dir + name_of_slice + '/' + 'analyzed_data/*/region_*/')
    cell_loc_path = glob.glob(thisdircheck + 'cell_metadata.csv')[0]
    countmatrix_path = glob.glob(thisdircheck+'cell_by_gene.csv')[0]
    sc_factor_1,sc_factor_2  = scaling_factor(main_dir)

    foo = pd.read_csv(cell_loc_path, index_col=0).sort_index()
    cell_pos_orig = foo[['center_x','center_y']]*np.array((sc_factor_1,sc_factor_2))
    cell_pos = foo[['center_x','center_y']]*np.array((sc_factor_1,sc_factor_2))/globalscaling
    adata_mer = run_denovo_cluster(countmatrix_path)
    
    save_filt_neuron_df(cell_pos_orig, adata_mer, 
                        outputfile=save_destination+f'/filt_neurons/filt_neurons_{name_of_slice}.csv', 
                       verbose=verbose)
    
    plot_jpg(cell_pos, adata_mer, dpi=100, save_bool=True,
             savefilename = save_destination+f'/image_xml/cell_img_{name_of_slice}.jpg',
            verbose=verbose)
    
    create_image_xml(name_of_slice, 
                 width=np.max(cell_pos.values, 0)[0], 
                 height=np.max(cell_pos.values, 0)[1], 
                 save_destination=save_destination, verbose=verbose)
    print(f'done for mouse: {name_of_slice}')
```
